```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, clear_output
```

```python
candy_data = pd.read_csv('/content/sample_data/candy.csv')
candy_data.head(5)
```

Popularity Between Candies


```python
# Filter the DataFrame
#more_popular = candy_data[candy_data['competitorname'].isin(['3 Musketeers', 'Almond Joy','Air Heads', 'Baby Ruth' ])]
#higher_sugar = candy_data[candy_data['competitorname'].isin(['Air Heads', 'Baby Ruth' ])]
#Scatterplot Setup
plt.figure(figsize=(30,20))
color =sns.color_palette()
sns.scatterplot(data=candy_data, x='competitorname', y='winpercent',size='sugarpercent',sizes=(50,300))
# Lines For The Answer
plt.axvline(x='Almond Joy', color='red', linestyle='-')
plt.axvline(x='3 Musketeers', color='red', linestyle='-')
plt.axvline(x='Baby Ruth', color='green', linestyle='-')
plt.axvline(x='Air Heads', color='green', linestyle='-')
# x and y Labels
plt.ylabel("Popularity")
plt.xlabel("Candy Name")
#Rotate the x axis label
plt.xticks(rotation=90)
# Clear the text output
clear_output(wait=True)
```

Correlation Between Popularity and Sugar Percent

```python
#Scatterplot Setup
plt.figure(figsize=(30,20))
sns.regplot(data=candy_data, x='sugarpercent', y='winpercent')
# x and y Labels
plt.ylabel("Popularity")
plt.xlabel("Sugar Percent")
#Rotate the x axis label
plt.xticks(rotation=90)
# Clear the text output
clear_output(wait=True)
```

Price relation between chocolate candies and non chocolate candies and it's popularity


```python
#lmplot Setup
sns.lmplot(data=candy_data, x='pricepercent', y='winpercent', hue='chocolate', height=10)
# x and y Labels
plt.ylabel("Popularity")
plt.xlabel("Price Percent")
#Rotate the x axis label
plt.xticks(rotation=90)
# Clear the text output
clear_output(wait=True)
```

Relation between chocolate and popularity

```python
#Scatterplot Setup
plt.figure(figsize=(30,20))
sns.swarmplot(data=candy_data, x='chocolate', y='winpercent', size=20)
# x and y Labels
plt.ylabel("Popularity")
plt.xlabel("Chocolate")
# Clear the text output
clear_output(wait=True)
```
